#### Условие 1: Задача 1
##### Считать данные с помощью pandas
##### Вывести на экран первые 5 строк
##### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
##### Если реновации дома не было, то в новом признаке поставьте 0
##### 1.3 Создайте признаки года продажи, месяца продажи
##### 1.4 Удалите признаки date, zipcode, lat, long

```python
import pandas as pd

# Чтение данных из файла и вывод первых 5 строк
data = pd.read_csv('kc_house_data.csv')
print(data.head())

# Создание признака delta_renovated
data['delta_renovated'] = data.apply(lambda row: row['yr_renovated'] - row['yr_built'] if row['yr_renovated'] > 0 else 0, axis=1)
print(data['delta_renovated'])
# Создание признаков года продажи и месяца продажи
data['year_sold'] = pd.to_datetime(data['date']).dt.year
data['month_sold'] = pd.to_datetime(data['date']).dt.month
print('year_sold = ',data['year_sold'],'month_sold = ',data['month_sold'])
# Удаление признаков date, zipcode, lat, long
data = data.drop(['date', 'zipcode', 'lat', 'long'], axis=1)
print(data)
```

```python
# 3.2
import numpy as np
df = pd.DataFrame(data)
# создаем сводную таблицу со столбцами condition и price
pivot_table = df.pivot_table(values='price', index='condition', 
                             aggfunc=[np.min, np.mean, np.max])

# присваиваем имена столбцам сводной таблицы
pivot_table.columns = ['Min Price', 'Mean Price', 'Max Price']

print(pivot_table)
```

```python
# 3.3
view_waterfront_table = pd.pivot_table(data=df, index=['view','waterfront'], columns=['grade'], values=['id'], aggfunc='count', fill_value=0)
print(view_waterfront_table)

```

```python
# 3.4
floors_bedrooms_table = pd.pivot_table(data=df, index=['floors', 'bedrooms'], values=['price'], aggfunc='count', fill_value=0)
print(floors_bedrooms_table)
```
